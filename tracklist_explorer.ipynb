```python
styles_per_episode = dict()
discarded = []
all_styles = []
for key in episodes:
    print(episodes[key])
    styles = []
    for track in episodes[key]:
        try:
            print(track[1], track[0])
            results = d.search(track[1], artist=track[0], type='release')
            styles.append(results.page(1)[0].styles)
        except:
            styles.append([])
            discarded.append([key, track[1], track[0]])
    styles_per_episode[key] = styles
    all_styles.append(styles)
```

```python
styles_per_episode
```

```python
styles_per_episode_flat = {}
```

```python
for key in styles_per_episode:
    episode_list = [i for i in styles_per_episode[key] if i]
    styles_per_episode_flat[key] = sum(episode_list, []) 
```

```python
styles_per_episode_flat
```

```python
counters = {}
for key in styles_per_episode_flat:
    temp = Counter(styles_per_episode_flat[key])
    counters[key] = Counter({k: c for k, c in temp.items() if c >=3 })
```

```python
for key in counters:
    plt.figure(figsize=(2, 2), dpi=80)
    plt.xticks(rotation='vertical')
    plt.bar(counters[key].keys(), counters[key].values())
```
