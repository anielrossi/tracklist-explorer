```python
all_lines = []

path = "files/"
savepath = "output.txt"
for files in glob.glob(path +"*.txt"):
    infile = open(files, 'r', encoding="utf8")
    lines = infile.readlines()
    for line in lines:
        if (re.findall(r'\d+', os.path.basename(files).split(".")[0])):
            row = line.strip().split("-") + re.findall(r'\d+', os.path.basename(files).split(".")[0])
        else:
            row = line.strip().split("-") + os.path.basename(files).split(".")[0].split("_")[1:]
            #print(line.strip().split("-") + "_".join(os.path.basename(files).split(".")[0].split("_")[1:]))
        all_lines.append(row)
    #a = infile.readline().split('.')
    #for k in range (0,len(a)):
    #    print(a[0], file=outfile, end='')
#outfile = open(savepath, 'w',  encoding="utf8")
#infile.close()
#outfile.close
all_lines
```

```python
def is_digit(row):
    return row['episode'].isdigit()
```

```python
df['personal'] = df.apply(lambda row : is_digit(row), axis=1)
```

```python
from collections import Counter

w = Counter(df['artist'])
```

```python
'n artists:' + str(len(w))
```

```python
mc = w.most_common(20)
mc
```

```python
q = Counter({k: c for k, c in w.items() if c >=3 })
```
