```python
episode_per_genre = dict()

for key in data.keys():
    g_flattened_per_episode = data[key]
    for j in g_flattened_per_episode:
        if j not in episode_per_genre:
            episode_per_genre[j] = []
        episode_per_genre[j].append(key)
```

```python
episode_per_genre_sorted = {key: value for key, value in sorted(episode_per_genre.items())}
```

```python
episode_per_genre_sorted
```

```python
from collections import Counter

counters = {}
for key in episode_per_genre_sorted:
    temp = Counter(episode_per_genre_sorted[key])
    counters[key] = Counter({k: c for k, c in temp.items() if c >= 3}).most_common(4)
```

```python
counters
```

```python
new_counters = {k:v for k,v in counters.items() if v}
```

```python
new_counters
```
